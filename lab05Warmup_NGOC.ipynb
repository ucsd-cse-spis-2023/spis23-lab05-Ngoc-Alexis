```python
from PIL import Image

bear = Image.open( "bear.png" )
```

```python
def invert( im ):
    ''' Invert the colors in the input image, im '''

    # Find the dimensions of the image
    (width, height) = im.size

    # Loop over the entire image
    for x in range( width ):
        for y in range( height ):
            (red, green, blue) = im.getpixel((x, y))
            # Complete this function by adding your lines of code here.
            # You need to calculate the new pixel values and then to change them
            red = 255 - red
            green = 255 - green
            blue = 255 - blue
            # in the image using putpixel()
            im.putpixel((x,y), (red,green,blue))
    #display(im)
invert(bear)
bear.save("new_bear.png") # create/overwrite new_bear.png with current image
display(bear)

```

```python
from PIL import Image


def invert_block(im):
    ''' Invert the colors in the input image, im '''

    # Find the dimensions of the image
    (width , height ) = im.size

    # Loop over the entire image
    for x in range(0, int(width * 0.25) ):
        for y in range(0, int(height * 0.25) ):
            (red, green, blue) = im.getpixel((x, y))
            red = 255 - red
            green = 255 - green
            blue = 255 - blue
            # in the image using putpixel()
            im.putpixel((x ,y), (red,green,blue))


# Load the image
bear = Image.open("bear.png")

# Invert the entire image
invert(bear)

# Invert the image in blocks
invert_block(bear)

# Save the modified image
bear.save("new_bear.png")

# Display the original and modified images
display(bear)
```
